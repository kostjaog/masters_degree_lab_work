```python
file_path = r'C:\Users\coolm\Downloads\Telegram Desktop\data\data16.txt'
```

```python
# Открываем файл для чтения
with open(file_path, 'r') as file:
    # Читаем строки из файла
    lines = file.readlines()

# Инициализируем два списка для значений X и Y
x_values = []
y_values = []

# Обрабатываем каждую строку
for line in lines:
    # Разделяем строки на две части (по пробелу) и преобразуем в числа с плавающей запятой
    x, y = map(float, line.split())
    
    # Добавляем значения в соответствующие списки
    x_values.append(x)
    y_values.append(y)

# Выводим результаты
print("Значения по оси X:", x_values)
print("Значения по оси Y:", y_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Путь к файлу
file_path = file_path

# Загружаем данные из файла в массив NumPy
data = np.loadtxt(file_path)

# Разделяем данные на значения X и Y
x_values = data[:, 0]
y_values = data[:, 1]

# Подбираем уравнение прямой методом наименьших квадратов
coefficients = np.polyfit(x_values, y_values, 1)

# Получаем уравнение прямой в виде строки
equation = np.poly1d(coefficients)

# Выводим уравнение прямой
print("Уравнение прямой:", equation)

# Строим точечный график
plt.scatter(x_values, y_values, label='Точки P(X, Y)')

# Строим аппроксимированную прямую
plt.plot(x_values, equation(x_values), color='red', label='Аппроксимированная прямая')

# Добавляем заголовки и метки осей
plt.title('Аппроксимация прямой к данным из файла')
plt.xlabel('X')
plt.ylabel('Y')

# Добавляем легенду
plt.legend()

# Отображаем график
plt.show()
```

```python
import numpy as np

# Путь к файлу
file_path = file_path

# Загружаем данные из файла в массив NumPy
data = np.loadtxt(file_path)

# Разделяем данные на значения X и Y
x_values = data[:, 0]
y_values = data[:, 1]

# Подбираем уравнение прямой методом наименьших квадратов
coefficients = np.polyfit(x_values, y_values, 1)

# Получаем уравнение прямой в виде строки
equation = np.poly1d(coefficients)

# Выводим уравнение прямой
print("Уравнение прямой:", equation)

# Вычисляем значения Y для аппроксимированной прямой
y_approx = equation(x_values)

# Вычисляем среднеквадратичное отклонение
mse = np.mean((y_values - y_approx)**2)

print("Среднеквадратичное отклонение:", mse)
```
